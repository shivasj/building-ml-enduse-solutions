# Explore Data

```python
# imports
import pandas as pd
import sqlite3
```

```python
# Connect to database
database_url = "../datastore/app_data.db"
database = sqlite3.connect(database_url)
```

```python
sql = "select * from articles"
articles = pd.read_sql_query(sql, database)

print("Shape:",articles.shape)
articles.head()
```

```python
# Counts by source
articles.groupby('source')["id"].count()
```

```python
# CNN
filtered = articles[articles["source"] == "cnn"]

filtered = filtered.sort_values(by=['article_dts'], ascending=False)
filtered[["article_date","article_title","article_content"]].head(10)
```
